## Subplots

````python
data = pd.read_csv('data-subplots.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages , py_salaries, label = 'Python')
plt.plot(ages, js_salaries, label = 'JavaScript')
plt.plot(ages, dev_salaries, label = 'All Devs' , color = '#444444', linestyle = '--')

plt.legend()

plt.title('Median Salary (USD) by Age')

plt.show()

````

The above plotting functions are not OOP friendly. We can make them more OOP friendly by using `subplots`

The way we are doing it now is called 'stateful', beacuse `plt` comes with a default state
We have been working with a single fure and single axis

```Python
plt.gcf() returns the current figure
plt.gca() returns the current axes ( Think of this as the plots )
```

**Figure** is a container for all the subplots
**Axes** is the plot in the container, think of it like the plots in the Figure


````python
fig, ax = plt.subplots()

# just replace `plt` with `ax`


ax.plot(ages , py_salaries, label = 'Python')
ax.plot(ages, js_salaries, label = 'JavaScript')
ax.plot(ages, dev_salaries, label = 'All Devs' , color = '#444444', linestyle = '--')

ax.legend()

# for title and legends we need to use the set_ should be used 
ax.set_title('Median Salary (USD) by Age')
ax.set_xlabel('Ages')
ax.set_ylabel('Median Salary (USD)')

plt.show()

````

````python
# by default it is 1x1
fig, ax = plt.subplots(nrows = 2, ncols = 2)

print(ax)

# See that the subplots are list of lists and not just list
````

````python
fig, (ax1, ax2) = plt.subplots(2, 1)

# just replace `plt` with `ax`


ax2.plot(ages , py_salaries, label = 'Python')
ax2.plot(ages, js_salaries, label = 'JavaScript')
ax1.plot(ages, dev_salaries, label = 'All Devs' , color = '#444444', linestyle = '--')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_xlabel('Ages')
ax1.set_ylabel('Median Salary (USD)')


ax2.legend()
ax2.set_title('Median Salary (USD) by Age')
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.show()

````

````python
# We dont need title for both 
# And we can also share the xaxis and yaxis

fig, (ax1, ax2) = plt.subplots(2, 1, sharex = True) # sharing x axis

# just replace `plt` with `ax`


ax2.plot(ages , py_salaries, label = 'Python')
ax2.plot(ages, js_salaries, label = 'JavaScript')
ax1.plot(ages, dev_salaries, label = 'All Devs' , color = '#444444', linestyle = '--')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')


ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.show()
````
