### Fills

```python

plt.style.use('fivethirtyeight')

plt.plot(py_dev_x, py_dev_y, label = 'Python')

plt.fill_between(py_dev_x, py_dev_y, alpha = 0.25)
# alpha tells about the transparency of the fill


plt.plot(dev_x, dev_y, color='#444444', linestyle='--',label = 'All Devs')

plt.title('Median Salary(USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary(USD)')

plt.legend()


plt.show()
```

```python
# make it more useful with conditional threshold

plt.style.use('fivethirtyeight')

plt.plot(py_dev_x, py_dev_y, label = 'Python')

overall_median = 57287

plt.fill_between(py_dev_x, py_dev_y, overall_median,alpha = 0.25)
# alpha tells about the transparency of the fill


plt.plot(dev_x, dev_y, color='#444444', linestyle='--',label = 'All Devs')

plt.title('Median Salary(USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary(USD)')

plt.legend()


plt.show()
```

```python
# conditional threshold with where

import numpy as np
py_dev_y = np.array(py_dev_y)

plt.style.use('fivethirtyeight')

plt.plot(py_dev_x, py_dev_y, label = 'Python')

overall_median = 57287

plt.fill_between(py_dev_x, py_dev_y, overall_median,
                    where = (py_dev_y > overall_median),
                    alpha = 0.25)
# alpha tells about the transparency of the fill


plt.plot(dev_x, dev_y, color='#444444', linestyle='--',label = 'All Devs')

plt.title('Median Salary(USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary(USD)')

plt.legend()


plt.show()
```

```python
# conditional threshold with where

import numpy as np
py_dev_y = np.array(py_dev_y)

plt.style.use('fivethirtyeight')

plt.plot(py_dev_x, py_dev_y, label = 'Python')

overall_median = 57287

plt.fill_between(py_dev_x, py_dev_y, overall_median,
                    where = (py_dev_y > overall_median),
                    alpha = 0.25)
# alpha tells about the transparency of the fill

plt.fill_between(py_dev_x, py_dev_y, overall_median,
                    where = (py_dev_y <= overall_median),
                    alpha = 0.25)

plt.plot(dev_x, dev_y, color='#444444', linestyle='--',label = 'All Devs')

plt.title('Median Salary(USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary(USD)')

plt.legend()


plt.show()
```

```python
# Fill between plots

import numpy as np
py_dev_y = np.array(py_dev_y)
dev_x = np.array(dev_x)
dev_y = np.array(dev_y)
py_dev_x = np.array(py_dev_x)

plt.style.use('fivethirtyeight')

plt.plot(py_dev_x, py_dev_y, label = 'Python')

overall_median = 57287

plt.fill_between(py_dev_x, py_dev_y, dev_y,
                    where = (py_dev_y > dev_y),
                    alpha = 0.25, label = 'Above Avg')
# alpha tells about the transparency of the fill

plt.fill_between(py_dev_x, py_dev_y, dev_y,
                    where = (py_dev_y <= dev_y),
                    alpha = 0.25, label = 'Below Avg')

plt.plot(dev_x, dev_y, color='#444444', linestyle='--',label = 'All Devs')

plt.title('Median Salary(USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary(USD)')

plt.legend()


plt.show()
```

```python
import pandas as pd
df = pd.read_csv('data.csv')
df.set_index('Responder_id', inplace = True)
df.head(5)
```

```python
from collections import Counter

language_counter = Counter()

for i,r in df.iterrows():
    language_counter.update(r['LanguagesWorkedWith'].split(';'))

data = language_counter.most_common(15)
```

```python
languages = list(zip(*data))[0]
popularity = list(zip(*data))[1]
```

```python
# lets plot
# But this time a horizontal bar plot using `barh`

plt.barh(languages, popularity)
plt.xlabel('Languages')
plt.ylabel('Popularity')

plt.show()
```

## Pie Charts

```python
# Basci Chart
plt.style.use('fivethirtyeight')

slices = [60, 40]
plt.pie(slices)


plt.title("Pie Chart")

plt.show()
```

```python
# adding labels
# Add wedge colors
# Add colours


slices = [120, 80, 30,20]
labels = ['Sixty', 'Forty', 'Twenty', 'Ten']
colors = ['blue', 'red', 'grey', 'brown']
plt.pie(slices, labels = labels , wedgeprops={'edgecolor':'black'}, colors = colors)


plt.title("Pie Chart")

plt.show()
```

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
```

```python
# explode a few slices
# add a shadow
# add a starting angle
# add the numbers using format trings

explode = [0, 0, 0, 0.1 , 0] # the vlaues tell the fraction of radius to explode

plt.pie(slices, labels = labels , wedgeprops={'edgecolor' : 'black'}, explode=explode, autopct = '%1.1f%%',shadow = True, startangle=90)
plt.show()
```

## Stack Plots

```python
minutes = list(range(1, 11))

player1 = [1, 2, 3, 4, 4, 4, 4, 4, 4, 5]
player2 = [1, 1, 1, 1, 2, 2, 2, 3, 4, 5]
player3 = [1, 2, 3, 4, 4, 4, 4, 4, 4, 5]

plt.stackplot(minutes, player1, player2, player3, labels = ['Player 1', 'Player 2', 'Player 3'])

plt.title('Stack Plot')
plt.legend(loc = 'upper left')

plt.show()
```
