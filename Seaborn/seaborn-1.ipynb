# Seaborn 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Histogram


- sns.distplot( data  , rug , kde , bins)

- sns.kdeplot( data , shade , rug  , bins)

### displot

```python
# data
x = np.random.normal(size = 1000)
```

```python
sns.set(color_codes = True)
```

```python
sns.displot(x)
plt.show()
```

Show the kde also by setting that property to be True.

```python
sns.displot(x, kde = True)
plt.show()
```

set the rug 

```python
sns.displot(x , kde = False, rug = True)
plt.show()
```

change the bins

```python
sns.displot(x, kde = False, rug = True, bins = 20)
plt.show()
```

### kdeplot

```python
sns.kdeplot(x)
plt.show()
```

Draw the shade also

```python
sns.kdeplot(x, shade = True)
plt.show()
```

## Box Plots

```python
sns.boxplot(data= x)
plt.show()
```

Set the position of whiskers

```python
sns.boxplot(data = x, whis = 0.2)
plt.show()
```

Set the orientation

```python
sns.boxplot(data = x, whis = 0.2, orient = 'h')
plt.show()
```

## Bar Plots

```python
# data
x = ['good', 'better', 'best', 'bad', 'worse', 'worst']
y = [1, 2, 3, 4, 5, 6]
```

```python
sns.barplot(x = x , y = y)
plt.show()
```
