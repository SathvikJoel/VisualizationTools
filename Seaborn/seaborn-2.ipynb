# Seaborn - 2

```python
import numpy as np 
import pandas as pd 

import seaborn as sns 
import matplotlib.pyplot as plt 

sns.set(color_codes = True)
```

## Jointplot

```python
# data

x = np.random.normal(size = 1000)
y = np.random.normal(size = 1000)
```

```python
sns.jointplot(x = x, y = y)
plt.show()
```

Add a kind and shade

```python
sns.jointplot(x = x , y = y , kind = 'kde', shade=True)
plt.show()
```

## Swarmplot

```python
# load a dataset
d = sns.load_dataset('diamonds')
```

```python
sns.swarmplot(x = d.sample(200).carat)
plt.show()
```

Use joint variables as well

```python
sns.swarmplot(x = 'cut', y = 'price', data = d.sample(200))
plt.show()
```

## Violinplot

```python
# data
p = sns.load_dataset('penguins')
```

```python
sns.violinplot(x = 'body_mass_g', data = p)
plt.show()
```

Multiple Violin Plots

```python
sns.violinplot(x = 'island', y='flipper_length_mm', data = p)
plt.show()
```

## Adv. Swarmplot

Add a hue to make it more useful

The color could also imply some data

```python
sns.swarmplot(x = 'island' , y = 'flipper_length_mm' , data= p , hue='sex' )
plt.show()
```

## Adv. Violinplot

make Paired Violin plots with hue

```python
sns.violinplot(x='island' , y='flipper_length_mm' , hue='sex' , data = p)
plt.show()

# The median , IQR etc... are being plotted for the entire data
```

Add a split to make it more useful

```python
sns.violinplot(x='island' , y='flipper_length_mm' , hue='sex' , split=True , data = p)
plt.show()
```

## Subplots

To draw subplots in the matplotlib OOP way use the ex argument for all the sns plots and the function returns the same ax object

```python
column_name = 'species'
nrows = len(p[column_name].unique())
fig , axs = plt.subplots(nrows =3);
for i , c_v in enumerate(p[column_name].unique()):
    pl = sns.kdeplot(p[p.species==c_v].flipper_length_mm , shade = True , ax=axs[i]);
    pl.set_title(c_v);
    
plt.tight_layout()
```
