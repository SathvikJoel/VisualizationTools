```python
import numpy as np 
import pandas as pd 

import seaborn as sns 
import matplotlib.pyplot as plt 

sns.set(color_codes = True)
```

```python
# data
p = sns.load_dataset('penguins')
d = sns.load_dataset('diamonds')
```

## Facet Grid

```python
g = sns.FacetGrid(p, row = 'species', col='sex')
g.map(sns.kdeplot, 'flipper_length_mm', shade = True)
plt.show()
```

## Pair plot

- Joint plot all the pairwise relationships in a dataset.

```python
sns.pairplot(p)
plt.show()
```

Add a `hue` and make it more meaningful.

useful when there is a categorical column

```python
sns.pairplot(p , hue = 'sex')
plt.show()
```

set corner to be True to avoid upper triangular elements

```python
sns.pairplot(p , hue = 'sex', corner=True)
plt.show()
```

## Boxen Plots

Plot other percentilies as well. Better version of box plots

```python
sns.boxenplot(x=d.sample(5000).carat)
plt.show()
```

Draw multiple boxen plots

```python
sns.boxenplot(x = 'island' , y = 'body_mass_g' , data = p)
plt.show()
```
