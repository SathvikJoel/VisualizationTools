```python
import numpy as np 
import pandas as pd 

import matplotlib.pyplot as plt 
from matplotlib import style
plt.style.use(['dark_background'])

import seaborn as sns 
sns.set(color_codes = True)
```

## Scatter plots

```python
# data
t = sns.load_dataset('tips')
t.head()
```

Add

- hue
- size
- style of the marker

```python
sns.scatterplot(x= 'total_bill' , y = 'tip' , data = t , hue = 'size' , style = 'sex');
```

## regplot

```python
# data
d = sns.load_dataset('diamonds')
```

```python
sns.regplot('x' ,'price' , data = d.sample(1000) , order = 2 , marker = '+');
plt.show()
```

## Bar plots

```python
sns.barplot(x = 'day' ,y = 'tip' , data = t);

# Split the data according to day
# Them Aggregating the mean 
# And plot the mean and also showing the confidence interval
```

```python
sns.barplot(x = 'day' ,y = 'tip' , data = t , estimator=np.median);
```

Add a hue

```python
sns.barplot(x = 'day' ,y = 'tip'  , hue = 'smoker', data = t , estimator=np.median);
```

- for scatter plot: It is continous variable(x) vs contionus variable(y)

- for Bar plot: It is contionus variable(y) vs Categorical Variable(x)

## Line Plot

```python
# data
f = sns.load_dataset('fmri')
```

```python
sns.lineplot(x = 'timepoint' , y = 'signal' ,hue = 'region', data = f);
```

Add one more layer of utility with style

```python
sns.lineplot(x = 'timepoint' , y=  'signal' ,hue = 'event', data = f , style = 'region');
```

## Heat Map

```python
# data
f1 = sns.load_dataset('flights')

f1.head()
```

```python
f1_ = pd.pivot(data = f1 , index = 'year', columns = 'month', values = 'passengers')
```

```python
sns.heatmap(f1_)
plt.show()
```

Add annotaions

```python
fig = plt.gcf()
fig.set_size_inches(12,8)
sns.heatmap(f1_.T , annot=True , fmt = 'd' , cmap = 'YlGnBu');
plt.show()
```
